## Linear regression with `linregress`

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Generate synthetic data with a negative linear trend
np.random.seed(42)
x = np.random.rand(100) * 10
y = -0.8 * x + np.random.normal(scale=1, size=x.shape)  # Negative linear correlation with some noise

# Use scipy to calculate the best-fit line
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Define the best-fit line
best_fit_line = slope * x + intercept

# Plot the scatter data
plt.scatter(x, y, label='Data', color='blue')

# Plot the best-fit line
plt.plot(x, best_fit_line, color='red', label=f'Best Fit Line (r={r_value:.2f})')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot with Best Fit Line (Negative Correlation)')
plt.legend()
plt.show()
```
